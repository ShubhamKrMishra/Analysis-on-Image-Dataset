```python
def Load_data(img_folder):
    #empty array to store the labels
    labels = []
    #empty array to store images
    img_data_array=[]
    for i in os.listdir(img_folder): #opening the directory
        for file in os.listdir(os.path.join(img_folder, i)): #searching for file in the specific directory
            image_path = os.path.join(img_folder, i,  file) #getting the path of the image
            img_data_array.append(image_path) #storing the image in the empty image array
            labels.append(i) # storing the corresponding label in the empty labels array
    return img_data_array, labels

data_image_path = Load_data("/content/drive/MyDrive/Shark Dataset Kaggle/sharks")
```

```python
import numpy as np
```

```python
datas = ['bull', 'blue', 'lemon', 'thresher', 'sand tiger', 'basking',
       'blacktip', 'hammerhead', 'mako', 'nurse', 'tiger', 'white',
       'whitetip', 'whale']
```

```python
data = []
labels = []

for category_idx, category in enumerate(categories):
    for file in os.listdir(os.path.join(input_dir, category)):
        img_path = os.path.join(input_dir, category, file)
        img = imread(img_path)  # img is a numpy array
        img= np.resize(img,(200,200,3))
        img = img.astype('float32')
        img /= 255 #scaling
        data.append(img.flatten())
        labels.append(category_idx)
```

```python
labels = np.asarray(labels)
data = np.asarray(data)
```

```python
data
```

```python
labels
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data,labels,test_size=0.2, stratify=labels)
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
# Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
# Decision tree Classifier

from sklearn.tree import DecisionTreeClassifier
dtc= DecisionTreeClassifier()
dtc.fit(x_train,y_train)
y_pred=dtc.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(x_train,y_train)
y_pred=rfc.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
# Decision tree Classifier

from sklearn.tree import DecisionTreeClassifier
dtc= DecisionTreeClassifier()
dtc.fit(x_train,y_train)
y_pred=dtc.predict(x_test)
accuracy_score(y_test,y_pred)
```
